<p align="justify">
Haz una gráfica para cada uno de las funciones evaluadas.
    
    

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,1000)
f_x = np.exp(-x**2)

y = x
X, Y = np.meshgrid(x, y)
f_z = Y*np.sin(X)


# Plot the surface
fig,ax = plt.subplots()
ax.plot(x,f_x)


fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, f_z)

plt.show()
```
