Debes ingresar las funciones matemáticas y evaluarlas en los vectores generados.

$$
f(x) =  e^{-x^2} 
$$

$$
f(z) =  ysin(x)

<p align="justify">
Haz una gráfica para cada uno de las funciones evaluadas.

El código Python utiliza NumPy y Matplotlib para generar y graficar dos funciones matemáticas. La primera función $ f(x) = e^{-x^{2}}$ se grafica en un gráfico de línea simple en función de $x$, mientras que la segunda función $( f(z) = y\sin(x) )$ se representa en un gráfico tridimensional con una superficie generada a partir de una malla de puntos $(X, Y)$. Se etiquetan los ejes y se proporcionan títulos para ambas gráficas, que se muestran al final del script.
    
    

```python
import numpy as np
import matplotlib.pyplot as plt

# Generacion de funcion f(x)
x = np.linspace(-5,5,1000)
f_x = np.exp(-x**2)

# Generacion de funcion f(z)
y = x
X, Y = np.meshgrid(x, y)
f_z = Y*np.sin(X)


# Grafica funcion f(x)
fig, ax = plt.subplots()
ax.grid(True, linestyle='-.')
ax.plot(x,f_x)
plt.xlabel("x")
plt.ylabel("$f(x)$")
plt.title("$f(x) = e^{-x^{2}}$")

# Grafica funcion f(z)
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, f_z,cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title("$f(z) = y\,sin({x})$")
plt.show()
```

### 3.2 Ruido Rosa
Investiga en que consiste el ruido rosa, cual es la diferencia del ruido gaussiano y demuéstralo.

El ruido rosa y el ruido gaussiano son dos conceptos relacionados con el análisis de señales y la ingeniería de sonido, pero tienen características diferentes:

* El ruido rosa es un tipo de señal aleatoria que tiene una densidad espectral de potencia constante por banda de octava. Esto significa que cada octava del espectro de frecuencias contiene la misma cantidad de energía.

* En términos prácticos, el ruido rosa suena como un zumbido constante y tiene más energía en las frecuencias más bajas que en las más altas.

* Se utiliza comúnmente en aplicaciones de ingeniería de sonido, terapia de sonido, pruebas de equipos electrónicos y como máscara de sonido.

El ruido gaussiano, también conocido como ruido blanco gaussiano, es un tipo de ruido aleatorio en el que la amplitud de las muestras sigue una distribución normal (gaussiana). A diferencia del ruido rosa, el ruido gaussiano no tiene una distribución de energía constante por banda de frecuencia. En otras palabras, puede haber más energía en ciertas frecuencias que en otras, y no sigue necesariamente un patrón específico como el del ruido rosa. El ruido gaussiano es ampliamente utilizado en diversos campos, incluyendo la electrónica, las comunicaciones y las ciencias de la computación, donde puede simular condiciones de ruido realista para propósitos de prueba y modelado. La principal diferencia entre el ruido rosa y el ruido gaussiano radica en cómo se distribuye la energía a lo largo del espectro de frecuencias: el ruido rosa tiene una densidad espectral de potencia constante por banda de octava, mientras que el ruido gaussiano no tiene una distribución de energía constante y sigue una distribución normal. 

### Comparacion del ruido gaussiano y rosa a travez de la PSD

El siguiente código de Python genera dos tipos de ruido, ruido gaussiano y ruido rosa, y luego calcula y grafica sus densidades espectrales de potencia (PSD) utilizando la biblioteca NumPy, Matplotlib y SciPy.

El código comienza generando dos tipos de ruido: ruido gaussiano y ruido rosa. El ruido gaussiano se crea utilizando una distribución normal de números aleatorios, mientras que el ruido rosa se obtiene filtrando el ruido gaussiano a través de un filtro Butterworth de segundo orden. Luego, se calcula la densidad espectral de potencia (PSD) para ambas señales utilizando el método de Welch, el cual divide la señal en segmentos para estimar el espectro de frecuencia. Esto proporciona información sobre cómo la energía de la señal está distribuida en diferentes frecuencias. Finalmente, se grafican las PSD para comparar visualmente las características espectrales del ruido gaussiano y el ruido rosa. Esta comparación ayuda a entender cómo cambian las propiedades de la señal después de aplicar el filtro Butterworth para generar el ruido rosa.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Generar ruido gaussiano
np.random.seed(0)  # Para reproducibilidad
gaussian_noise = np.random.normal(0, 1, 1000)  # Media 0 y desviación estándar 1

# Generar ruido rosa utilizando un filtro de coloración
b, a = signal.butter(2, 0.1, 'low', analog=False)
pink_noise = signal.filtfilt(b, a, gaussian_noise)

# Calcular PSD utilizando la transformada de Fourier
frequencies_gaussian, psd_gaussian = signal.welch(gaussian_noise, fs=1.0, nperseg=1024)
frequencies_pink, psd_pink = signal.welch(pink_noise, fs=1.0, nperseg=1024)

# Graficar PSD
plt.figure(figsize=(10, 6))
plt.semilogy(frequencies_gaussian, psd_gaussian, label='Ruido Gaussiano', color='blue')
plt.semilogy(frequencies_pink, psd_pink, label='Ruido Rosa', color='red')
plt.title('Densidad Espectral de Potencia (PSD)')
plt.xlabel('Frecuencia')
plt.ylabel('PSD')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
